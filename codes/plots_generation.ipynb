```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
macro = [[0.2772, 0.2807, 0.1899,0.1848],
         [0.2851, 0.2297, 0.1849, 0.1735],
         [0.4086, 0.3243, 0.2483, 0.2481],
         [0.3791, 0.3094, 0.2331, 0.2434]]
```

```python
macro_np = np.array(macro)
```

```python
x = ['Indic-bert', 'Indic-bert-lat', 'Multilingual-bert', 'Multilingual-bert-lat']
```

```python
# Plotting the lines
plt.plot(x, macro_np[:,0], marker='*',label='Urdu')
plt.plot(x, macro_np[:,1],marker='*', label='Hindi')
plt.plot(x, macro_np[:,2],marker='*', label='Tamil')
plt.plot(x, macro_np[:,3],marker='*', label='Telugu')

# Adding labels and title
plt.xlabel('Models')
plt.ylabel('Macro F1')
# plt.title('Language-wise comparison across different models (Macro F1)')

# Adding legend
plt.legend()
plt.savefig('Macro_f1.png', dpi=300)  # You can specify the file format, such as .png, .jpg, .pdf, etc.

# Displaying the plot
plt.show()

```
